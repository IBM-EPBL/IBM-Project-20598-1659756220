```python
#Here I used an excel sheet which contain data of the age with the given condition which is adding 1.5 to the Rings gives the age 
import pandas as pd
ds=pd.read_excel("Abalone_age_prediction.xlsx")
```

```python
#dependent variables
x=ds.iloc[:,1:9]
x
```

```python
#independent variable
y=ds.iloc[:,9:10]
y
```

```python
#Train,Test & Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Model Building
from sklearn.linear_model import LinearRegression 
slr=LinearRegression()
slr.fit(x_train,y_train)
```

```python
slr.predict(x_test)
```

```python
#Measure the performance using Metrics
from sklearn.metrics import r2_score
r2_score(slr.predict(x_test),y_test)
```
